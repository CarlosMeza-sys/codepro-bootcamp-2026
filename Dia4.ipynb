```python
class Superheroe:
    # Atributo de clase
    especie = 'superhumanos'

    # Definir el metodo constructor para definir atrib. de instacia
    def __init__(self, nombre):
        self.nombre = nombre
    
    # Otros Metodos
    def rayos_laser(self):
        print('Tiro rayos laser rojos')

    def volar(self, km):
        print(f'Estoy volando a {km}')

    def alzar_objetos(self, objeto, kg):
        print(f'Puedo alzar un {objeto} de {kg} kg')

# Instanciar objetos
superheroe1 = Superheroe('Batman')
superheroe1.volar(120)

superheroe2 = Superheroe('Flash')
superheroe2.alzar_objetos('trenes', '500')
```

```python
class Coche:
    tipo = 'vehiculo'

    def __init__(self, modelo, posicion = 0):
        self.modelo = modelo
        self.posicion = posicion
    
    def acelerar(self, metros):
        self.posicion += metros
        print(f'El vehiculo modelo {self.modelo} acelera y avanza {metros}. Nueva posicion: {self.posicion}m')

    def girar(self, direccion):
        print(f'El {self.modelo} avanza y gira a la direccion {direccion}')

coche1 = Coche('Toyota')
coche1.acelerar(1)
coche1.girar('izquierda')
```

```python
class Mascota:
    tipo = 'animal'

    # Metodo constructor
    def __init__(self, raza, color):
        self.raza = raza
        self.color = color
    
    # Metodos
    def hablar(self, habla):
        print(f'Cuando hablo digo {habla}')

    def mover(self, mover):
        print(f'Cuando me muevo hago {mover}')

mascota1 = Mascota('bulldog','negro')
mascota1.hablar('hola')
mascota1.mover('un salto')
```

```python
class AnimalDeGranja:
    # Atributo de clase:
    tipo = 'animal de granja'

    # Atributos de instacia
    def __init__(self, nombre, color, sonido):
        self.nombre = nombre
        self.color = color
        self.sonido = sonido

    # Metodo (Acciones de la clase)
    def hablar(self):
        print(f'Hablo y digo {self.sonido}')

    def presentarse(self):
        print(f'Hola, soy {self.nombre}')

class Gallina(AnimalDeGranja):
    huevos_por_dia = 1

    def poner_huevo(self):
        print(f'Estoy poniendo {self.huevos_por_dia}')

class Vaca(Gallina):
    patas = 4

    def mujir(self):
        print('Muuu')

gallina1 = Gallina('Hortencia', 'marron', 'corococo')
gallina1.hablar()
gallina1.presentarse()

vaca1 = Vaca('Lola', 'blanco y negro', 'Muuu')
vaca1.hablar()
vaca1.presentarse()
vaca1.patas
```

```python
class Mascota:
    tipo = 'animal domestico'

    def __init__(self, nombre, color, comida, especie):
        self.nombre = nombre
        self.color = color
        self.comida = comida
        self.especie = especie

    def presentar_color(self):
        print(f'Soy {self.nombre}, un {self.especie} de color {self.color}')

    def decir_comida(self):
        print(f'Mi comida favorita es {self.comida}')

class Perro(Mascota):
    cant_de_patas = 4

    def decir_patas(self):
        print(f'Tengo {self.cant_de_patas} patas')

class Gato(Perro):
    velocidad = 'rapido'

    def correr(self):
        print(f'Cuando salgo corriendo, lo hago {self.velocidad}')

# Instaciar objetos
perro1 = Perro('Firulais', 'marron', 'hueso', 'perro')
perro1.presentar_color()
perro1.decir_comida()

gato1 = Gato('Alfred', 'negro', 'comida para gato', 'gato')
gato1.presentar_color()
gato1.decir_comida()
```
